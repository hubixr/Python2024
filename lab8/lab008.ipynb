```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

```python
# Definicja zmiennych i funkcji
x = sp.Function('x')
t = sp.Symbol('t', real=True)
w0 = sp.Symbol('omega_0', real=True, positive=True)
v0 = sp.Symbol('v_0', real=True)
x0 = sp.Symbol('x_0', real=True)
w = sp.Symbol('omega', real=True)
F0 = sp.Symbol('F_0', real=True)
gamma = sp.Symbol('gamma', real=True, positive=True)

# Równanie oscylatora tłumionego
osc_t = sp.Eq(x(t).diff(t, t) + 2 * gamma * x(t).diff(t) + w0**2 * x(t), 0)
sol_tl = sp.dsolve(osc_t, x(t), ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})
```

```python
# Wykres 
sol = sol_tl.rhs.subs({x0: 0, w0: 1, v0: 1, gamma: 0.1})
sp.plot(sol, (t, 0, 40), title="Oscylator tłumiony")
```

```python
# Wykres
sol = sol_tl.rhs.subs({x0: 0, w0: 1, v0: 1, gamma: 1.1}).evalf()
sp.plot(sol, (t, 0, 40), title="Oscylator tłumiony")
```

```python
# Wykres
sol = sol_tl.rhs.subs({x0: 0, w0: 1, v0: 1, gamma: 0})
sp.plot(sol, (t, 0, 40), title="Oscylator tłumiony")
```

```python
#Oscylator wzbudzony
a0=sp.Symbol('alpha_0', real=True)
# Równanie oscylatora tłumionego z wymuszeniem
osc_wzbudzony = sp.Eq(x(t).diff(t, t) + 2 * gamma * x(t).diff(t) + w0**2 * x(t), F0 * sp.cos(w * t))

# Rozwiązanie ogólne z warunkami początkowymi
sol_wzbudzony = sp.dsolve(osc_wzbudzony, x(t), ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})

# Wyświetlenie rozwiązania
display(sol_wzbudzony)

# Wykres
sol = sol_wzbudzony.rhs.subs({x0: 0, v0: 0, w0: 1, gamma: 0.1, F0: 1, w: 1})
sp.plot(sol, (t, 0, 40), title="Oscylator wzbudzony")
```

```python
sol = sol_wzbudzony.rhs.subs({x0: 0, v0: 0, w0: 1, gamma: 0.05, F0: 1, w: 0.95})
sp.plot(sol, (t, 0, 100), title="Rezonans – słabe tłumienie, ω ≈ ω_0")
```
